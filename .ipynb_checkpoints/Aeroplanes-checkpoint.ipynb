## Predicting Flight Delays:

### Importing, Cleaning and Introduction:

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn import ensemble
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('2008_aeroplanes.csv')
df.shape
```

```python
null_vals = pd.DataFrame(df.isna().sum())
display(null_vals)
```

> As there are a significant amount of null values within this dataset, dropping all of it will remove a significant amount of meaning from the dataset.  

> Given that some of the columns with null values regarding arrival/departure time are relatively small in numbers (i.e. under 150,000 rows) and deleting rows with such values would only represent dropping approximately 7% of the data, we will go ahead and do that for now.  

> It is interesting to note that, by only dropping rows with null values in the seven rows below certainly cleans up a lot of other null values in other columns.  This could point to some sort of bias (i.e. companies that haven't recorded arrival time or taxing times were not recording other information (perhaps to cover up delays or perhaps for some other reason that might be worth investigating).  After totalling up the columns that we dropped after the code below, only 154699 columns were dropped (which is approximately 2% of the data).  Thus, you would probably only have a situation with bias if the majority of these rows were either under the same airline or involved the same plane or arrival/destination city.

```python
df = pd.DataFrame(df.dropna(subset = ['DepTime', 'ArrTime', 'ActualElapsedTime', 'CRSElapsedTime',
                                      'AirTime', 'ArrDelay', 'DepDelay'], how = 'any'))
```

### Exploration:

#### General Overview:

```python
# Correlation heatmap:

f, ax = plt.subplots(figsize = (17, 15))
sns.heatmap(df.corr(), annot = True)
```

> From this heatmap, we can see that a Carrier Delay and a Late Aircraft Delay have a relatively high correspondence with both arrival and departure delays.  NAS Delays (National Air System) have a significant effect on the time between leaving the departure gate and being in the air towards your destination city (i.e. TaxiOut).

> We still have to consider locations and carriers, however - as these represent categorical data and not continuous (and therefore fall outside of our correlation map capabilities).
