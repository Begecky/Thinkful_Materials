## Global Terrorism Database (Focusing on Cyber-Related Terrorism):

This dataset is from Global Terrorism Database curated by the University of Maryland:

National Consortium for the Study of Terrorism and Responses to Terrorism (START). (2018). Global Terrorism Database [Data file]. Retrieved from https://www.start.umd.edu/gtd.

In dividing and classifying the city names within the data-set, __[this article from the New York Times](https://www.nytimes.com/interactive/2016/01/04/world/middleeast/sunni-shiite-map-middle-east-iran-saudi-arabia.html)__ was instrumental as was __[this website](http://zakhor-online.com/?attachment_id=7932)__ and __[this one.](https://www.vox.com/a/maps-explain-the-middle-east)__
__[and this one too](http://gulf2000.columbia.edu/maps.shtml)__ https://www.crisisgroup.org/latin-america-caribbean/andes/colombia/63-colombias-armed-groups-battle-spoils-peace#map-5700-8

In this notebook, we will try and isolate data related to cyber-warfare or hacking events.

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### 1. Importing, Cleaning and General Overview:

```python
df = pd.read_excel('globalterrorismdb_0718dist.xlsx', 
                   usecols = 'A:D, F, H:M, S:W, AA:AD, AI:AN, AP, BG, BM:BN, BQ:BR, CD:CG, DA',
                   dtype = {'summary':str, 'motive':str})                                                                                         
```

```python
# Renaming for usability:

df = pd.DataFrame(df.rename({'eventid':'event_id', 'iyear':'year', 'imonth':'month',
             'iday':'day', 'extended':'24+hrs', 'provstate':'region_2', 'doubtterr':'doubt', 
             'attacktype1':'attack_1', 'attacktype1_txt':'attack_1txt','attacktype2':'attack_2', 
             'targtype1':'target_1', 'targtype1_txt':'target_1txt', 'targsubtype1':'sub_target',
             'targsubtype1_txt':'sub_targettxt', 'target1':'specific_target',
             'natlty1':'victim_nationality', 'natlty1_txt':'victim_nationalitytxt', 'gname':'group_name', 
             'guncertain1':'group_attrib_crtainty', 'individual':'unaffil_individ', 'weaptype1':'weapon', 
             'weaptype1_txt':'weapontxt', 'weapsubtype1':'sub_weapon', 'weapsubtype1_txt':'sub_weapontxt'}, axis = 1))
```

```python
# Looking at the size and amount of null values in our dataset:

pd.set_option('max_rows', 47)

print(df.isna().sum())
print(df.shape)
```

```python
# Cleaning up some of the null values in our object-type columns:
# (not dropping them but setting them to 'No Data'):

df.motive.where(df.motive.apply(lambda x: x != 'nan'), other = 'No Data')
df.summary.where(df.summary.apply(lambda x: x != 'nan'), other = 'No Data')
```

### 2. Preliminary Exploration and Visualization:

##### A. Isolating relevant data - Regex:

> Here, we want to try to isolate the rows whose summaries relate directly to what we want to analyze:

```python
# Creating the dyber_data1 Dataframe:
# Regex to isolate qualifying data (This set is more flexible in its Regex):

import re

print(df.summary.str.contains('cyb+|social|inform+|network+|twit+|facebook|fake|electio+|internet|ads|adv+|priv+|hac+|breac+|emai+',
                         flags = re.IGNORECASE, regex = True).value_counts())
```

```python
# Creating a mask which represents the results from our search above:

mask = df.summary.str.contains(r'(cyb+|social|inform+|network+|twit+|facebook|fake|electio+|internet|ads|adv+|priv+|hac+|breac+|emai+)', 
                           flags = re.IGNORECASE)

# Creating a new, filtered dataframe with this specific information:

cyber_data1 = pd.DataFrame(df[mask])
starting_rowcount = cyber_data1.shape[0]

# Using cyber_data1 to make a df of the 30 countries with the most instances of cyber-terrorism:

top_30_countries = cyber_data1.country_txt.value_counts().head(30)
mask2 = cyber_data1.country_txt.map(lambda x: x in top_30_countries)
cyber_top30 = cyber_data1[mask2]
```

### 3. Wranglin':

##### Dropping inconsequential NA Values:

```python
# Since making dummies of the entire dataframe didn't quite work, looking at doing some customized dummy columns
# and then using the get_dummies method for the rest:

# Running into NA issues, so dropping na where na values are under 50:

cyber_data1 = pd.DataFrame(cyber_data1.dropna(subset = ['city', 'group_attrib_crtainty',
                                                        'specific_target']))
```

##### Wrangling the more considerable nulls:

```python
# Taking out the NA the mean of the remaining values and 
# setting the values we did not consider with the mean of the others: 

# sub_target:

sub_target_mean = cyber_data1.sub_target.mean(skipna = True)

cyber_data1.sub_target.fillna(sub_target_mean, inplace = True)

# sub_targettxt:
# Following the same pattern as above:

cyber_data1.sub_targettxt.fillna('No Data')

# victim_nationality:

cyber_data1.victim_nationalitytxt.fillna('No Data')

# sub_weapon and sub_weapontxt:
# Filling in the NA with the mean:

sub_weapon_mean = cyber_data1.sub_weapon.mean(skipna = True)

cyber_data1.sub_weapon.fillna(sub_weapon_mean, inplace = True)

# Filling the NA with 'No Data'

cyber_data1.sub_weapontxt.fillna('No Data')
```

#### Manual Dummies:

```python
print(cyber_data1.shape)

# Still a lot of unique values - will try dropping the ones with unmanageable numbers for our model.
# Will try grouping with the others

cyb1_dscr_ob = cyber_data1.describe(include = [np.object])

cyb1_dscr_num = cyber_data1.describe()
```

##### weapon and weapon_txt:

```python
# Weapon:
# Grouping related values together after looking at value-counts:

# Grouping Chemical and Biological Weapons together:
cyber_data1['weapon'] = cyber_data1['weapon'].replace(1, 2)
# Putting Fake and 'Other' types of weapons into the same category:
cyber_data1['weapon'] = cyber_data1['weapon'].replace(12, 7)

# Putting vehicles into category 11 (sabotage equipment):
cyber_data1['weapon'] = cyber_data1['weapon'].replace(10, 11)
```

```python
# Weapon_txt:

# Grouping Biological and Chemical Weapons:
cyber_data1['weapontxt'] = cyber_data1['weapontxt'].replace('Chemical', 'Bio-Chem')
cyber_data1['weapontxt'] = cyber_data1['weapontxt'].replace('Biological', 'Bio-Chem')

# Grouping Fake and Other together:
cyber_data1['weapontxt'] = cyber_data1['weapontxt'].replace('Fake Weapons', 'Fake/Other')
cyber_data1['weapontxt'] = cyber_data1['weapontxt'].replace('Other', 'Fake/Other')

# Aggregating Vehicle into Sabotage Equipment:
cyber_data1['weapontxt'] = cyber_data1['weapontxt'].replace(
    'Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)', 'Sabotage Equipment')
```

##### sub_target and sub_targettxt

```python
# Sub_target:

# I didn't want to isolate all of this, else it would reflect the target column exactly - 
# but made some sub-groupings:

industrial = [1, 5, 9, 10, 12]
white_collar = [2, 3, 4, 6, 7, 8, 11, 13, 112]
gov_figure1 = [14, 18]
gov_figure2 = [15, 16, 17, 19]
police_facilities = [22, 23, 24, 26]
mil_facilities = [27, 28, 30, 31, 32, 35, 36, 39]
mil_personnel = [29, 33, 34]
pilots_planes = [42, 43]
gov_diplomatic = [45, 46, 47]
educational = [48, 49, 50]
food_water = [51, 52]
media = [53, 54, 55, 56]
religious = [69, 85, 86, 87]
political = [83, 84, 109, 110, 111]
mass_socio = [65, 68, 70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 81, 82, 95, 96, 97, 98, 99, 100, 101, 102, 103, 104, 105]
util_telecom = [88, 89, 90, 91, 92, 106, 107, 108]
```

```python
# Sub_target Assignment:

cyber_data1['sub_target'].replace({'sub_target': {'industrial':1, 'white_collar':2, 
                                       'gov_figure1':14,'gov_figure2':15, 
                                       'police_facilities':22, 'mil_facilities':27,
                                       'mil_personnel':29, 'pilots_planes':42, 
                                       'gov_diplomatic':45, 'educational': 48, 
                                       'food_water':51, 'media':53, 'religious':69,
                                       'political':83, 'mass_socio': 65,
                                       'util_telecom': 88}}, inplace = True)
```

```python
# Sub_targettxt:

# Regex:

ind = re.compile(r'(?is)(Gas\/Oi.*)|(Industrial\/Tex.*)|(Farm\/Ranch)|(Mining)|(Construction)')
wc = re.compile(r'''(?isx)(Restaurant\/Ba.*)|(Bank\/Commerce)|(Multinational.*)|(Medical\/Phar.*)|
                          (Retail\/Groc.*)|(Hotel\/Resort)|(Entertainment\/Cultural.*)|(Private\sSecuri.*)|
                          (Legal\sServices)''')
gov_1 = re.compile(r'(?is)(Judge\/Atto.*)|(Government\sPerson.*)')
gov_2 = re.compile(r'(?is)(Politician\sor\sPoli.*)|(Royalty)|(Head\sof\sState)|(Election-related)')
police_fac = re.compile(r'(?is)(Police\sBuildings\s.*)|(Police\sPatr.*)|(Police\sCheckpoint)|(Prison\/Jail)')
mil_fac = re.compile(r'''(?isx)(Military\sBar.*)|(Military\sRecr.*)|(Military\sWeap.*)|(Military\sAir.*)|
                         (Military\sMari.*)|(Military\sTransportation\/Vehic.*)|(Military\sCheckpo.*)|
                         (Paramilitary)''')                  
mil_pers = re.compile(r'(?is)(Military\sUnit\/Patro.*)|(Non-combatant\sPers.*)|(Military\sPersonne.*)')
pilots = re.compile(r'(?is)(Aircraft.*)|(Airline\sOffic.*)')
gov_dip = re.compile(r'(?is)(Diplomatic\sPersonn.*)|(Embassy\/Consulate)|(International\sOrgan.*)')
edu = re.compile(r'(?is)(Teacher\/Profe.*)|(School\/Univer.*)|(Other\sPersonnel)')
food = re.compile(r'(?is)(Food\sSupply)|(Water\sSupply)') 
med = re.compile(r'''(?isx)(Newspaper\sJournalist\/Sta.*)|(Radio\sJournalist\/Sta.*)|(Television\sJournalis.*)|
                           (Other\s\(including\sonline\snews\sagencies\))''')
relig = re.compile(r'''(?isx)(Religion\sIdentified.*)|(Religious\sFigure)|(Place\sof\sWorship)|
                       (Affiliated\sInstitution)''')
polit = re.compile(r'(?is)(Protester)|(Political\sParty\sMem.*)| (Party\sOfficial\/Candi.*)|(Party\sOff.*)|(Rally)')
m_socio = re.compile(r'''(?ix)(Refugee\s.*)|(Student)|(Race\/Eth.*)|(Farmer)|(Vehicles\/Tran.*)|(Marketpla.*)|
                             (Village\/City\/To.*)|(House\/Apartm.*)|(Laborer\s.*)|(Procession\/Gat.*)|
                             (Public\sArea\s.*)|(Memorial\/Ceme.*)|(Museum\/Cul.*)|(Labor\sUnion.*)|
                             (Tourism\sTrav.*)|(Tour\sBus.*)|(Tourist)|(Other\sFacility)|(Bus\s.*)|
                             (Train\/Tra.*)|(Bus\sStat.*)|(Subway)|(Bridge\/C.*)|(Highway\/Ro.*)|(Taxi\/Ric.*)''')
util = re.compile(r'''(?isx)(Radio)|(Television)|(Telephone/Tele.*)|(Internet\sInfr.*)|(Multiple\sTelecommu.*)|
                            (Gas|Electricity|Oil)''')
```

```python
# Sub_targettxt Assignment:

cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(ind, 'Industrial')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(wc, 'White_Collar')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(mil_pers, 'Mil_Personnel') 
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(gov_1, 'Gov_Figure1')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(gov_2, 'Gov_Figure2')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(police_fac, 'Police_Facilities')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(mil_fac, 'Mil_Facilities') 
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(edu, 'Educational')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(pilots, 'Pilots_Planes')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(gov_dip, 'Gov_Diplomatic')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(food, 'Food_Water')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(med,'Media')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(relig, 'Religious')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(polit, 'Political')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(m_socio, 'Mass_Socio')
cyber_data1['sub_targettxt'] = cyber_data1['sub_targettxt'].str.replace(util, 'Util_Telecom')
```

### 4. First Models:

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import mean_squared_error, confusion_matrix
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
# Trying the cyber_data1 dataset:

# Defining our input and output:

X = pd.get_dummies(cyber_data1[['year', 'month', '24+hrs', 'country_txt','region_txt', 
                 'crit1', 'crit2', 'crit3', 'doubt', 'suicide', 'attack_1txt',
                 'target_1txt', 'sub_targettxt', 'victim_nationalitytxt', 'group_name',
                 'group_attrib_crtainty', 'unaffil_individ', 'weapontxt', 'sub_weapontxt', 'property']])
        
y = cyber_data1['success']

# Splitting the data:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25)
```

```python
# Fitting the model:
    
forest_cl = RandomForestClassifier()
forest_cl.fit(X_train, y_train)
y_pred_ = forest_cl.predict(X_test)
    
# Evaluation:
    
confusion_forest = confusion_matrix(y_test, y_pred_, labels = [0, 1])
forest_cv = cross_val_score(forest_cl, X_train, y_train, cv = 5)
forest_train_score = forest_cl.score( X_train, y_train)
forest_test_score = forest_cl.score( X_test, y_test)
    
```

```python
cyber_data1.success.value_counts()
```

```python
plusminus = u"\u00B1"

print("The training score is: {:.2%}".format(forest_train_score))
print("\nThe mean cross-validation score is: {:.2%} {}{:.2%}\n".format(forest_cv.mean(),
                                                                      plusminus, forest_cv.std()))
print("The cv scores are: \n{}".format(forest_cv))
print("\nThe test score is: {:.2%}".format(forest_test_score))

forest_confusion_df = pd.DataFrame(confusion_forest)

FP = forest_confusion_df.loc[0, 1]
FN = forest_confusion_df.loc[1, 0]
NegT = forest_confusion_df.iloc[0].sum()
PosT = forest_confusion_df.iloc[1].sum()

print(forest_confusion_df)

print('False Positive (Type I Error): {}\n({:.2%})\n'.format(FP, (FP / NegT)))
print('False Negative (Type II Error): {}\n({:.2%})\n'.format(FN, (FN / PosT)))

feat_imp = pd.DataFrame(forest_cl.feature_importances_.round(2), index = X_train.columns, columns = ["Importances"])
display(feat_imp.sort_values('Importances', ascending = False))
```
