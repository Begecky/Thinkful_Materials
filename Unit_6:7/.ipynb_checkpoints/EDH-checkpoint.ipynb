```python
# Variables for upcoming loops / comparisons:

ctry_list = list(country_dict.keys())
url_list = list()

print(len(ctry_list))
```

```python
# Generating our API queries:

for x in ctry_list:
    
    url_query = 'https://edh-www.adw.uni-heidelberg.de/data/api/fotos/search?country={}'.format(x)
    url_list.append(url_query)
```

```python
# Making sure it came through:

print(len(url_list))
```

```python
# Splitting this into smaller lists (responsible API-ing):

A = url_list[0:16]
B = url_list[16:31]
C = url_list[31:46]
D = url_list[46:60]

# A list containing the urls we removed for easy reference:
zeros = list()

# Our final list:
final_list = list()
```

```python
def url_looper(listed_urls):
    
    # All-purpose counter McThingy:
    counter = 0
    
    # First swig o' the data:
    for url in listed_urls:
        
        response = requests.get(url)
    
        response_dict = response.json()
        count = response_dict.get('total')
    
        if count == 0:
            listed_urls.remove(url)
            zeros.append(url)
    
        else:
            url = url+'&limit={}'.format(count)
            final_list.append(url)
        
        time.sleep(15)
        print(counter)
        counter += 1
```

```python
url_looper(A)
```

```python
main_df = pd.DataFrame()
counter_2 = 0
```

```python
for url in final_list:

    x = requests.get(final_list[counter_2])
    x_json = x.json()
    
    bits = x_json.get('items')
    
    for x in bits:
        main_df = main_df.append(x, ignore_index = True, sort = True)
        
    counter_2 += 1
    print(counter_2)
```
